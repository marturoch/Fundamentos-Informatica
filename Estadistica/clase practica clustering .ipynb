```python
import pandas as pd
datos = pd.read_csv("https://raw.githubusercontent.com/AJVelezRueda/Fundamentos_de_informatica/master/Ciencia_de_datos/practicos/recursos/Datos.csv", sep=",")
datos
```

```python
datos.info()
```

```python
datos.describe()
```

```python
#Localizamos los outliers
datos.plot.scatter(x='Sueldo', y='Altura', figsize = (10, 5))
```

```python
#Filtrado de outliers
datos = datos[(datos.Sueldo < 200000) & (datos.Altura < 10)]
datos
```

```python
#Verificamos el filtrado
datos.plot.scatter(x='Sueldo', y='Altura', figsize = (10, 5))
```

```python
datos.info()
#Antes habian 106 observaciones, y haciendo el filtrado de los outliers ahora hay 98.
```

```python
#Eliminamos las celdas repetidas
datos.drop_duplicates(inplace=True)
#datos.drop_duplicates(subset="Altura", inplace=True)
```

```python
#Sacando los outliers teniamos 98 observaciones. Sacando los duplicados tenemos 94 observaciones. De estas 94 no hay datos nulos.
datos.info()
```

```python
datos[['Sueldo','Altura']].describe() #analisis estadistico
#Antes la media era 1.94 y ahora 1.72 
#el desvio era 1.56 y ahora 0.28 
#El max era 13.7 y ahora 2.35
#Ahora tiene mas sentido
```

```python
datos[['Sueldo','Altura']].median()
#La mediana de sueldo no es cercana a la media 
#La mediana de altura se acerca a la media -> distribucion normal
```

```python
datos.Altura.mode()
```

```python
#Como no hay datos repetidos me muestra todos
datos.Sueldo.mode()
```

```python
'''
Analisis

Sueldo: El promedio(94k) y la mediana(99k) estan muy alejados. Otro dato que nos proporciona esto es el desvio estandar, el cual es 2400; es un desvio muy alto. ¿Que nos dice esto? Como sabemos una distribucion normal la media es igual a la mediana, por ende podemos deducir que nuestra distribucion no se comporta como la de una normal.

Altura: La media es 1.72 y la mediana 1.75, con un desvio de .27. No es una distribucion normal perfecta, pero si podemos decir que se acerca y se comporta como una.

A continuacion vemos esto graficamente.

Con scipy se puede verificar la normalidad de una distribucion, mas adelante lo demostramos.
'''
```

```python
import seaborn as sns
sns.histplot(data = datos, x = "Altura", binwidth=0.25, kde = True)
#comportamiento muy aproximado a una distribucion normal
```

```python
import seaborn as sns
sns.histplot(data = datos, x = "Sueldo", kde = True)
#No solo no es una dsitribucion normal sino que al ver 2 picos, esto es comun cuando unimos 2 curvas distintas(es como 2 normales solapadas)
```
