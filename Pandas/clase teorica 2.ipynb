```python
#1. eliminaci√≥n de casos completos, es decir eliminar toda las filas 
#que contienen un dato faltante:
#creamos dataframes para probar los filtros, 
clean = pd.DataFrame(personas)
clean1 = pd.DataFrame(personas)
clean2 = pd.DataFrame(personas)
clean3 = pd.DataFrame(personas)
clean4 = pd.DataFrame(personas)
clean.dropna(inplace=True) #elimina las filas donde por lo menos hay un valor nulo
clean.info()
```

```python
#Perdida de datos(entradas, filas)
#Perdi el 30%
100 - (len(clean.columns) * 100) / len(personas.columns) #0.0
100 - (len(clean) * 100) / len(personas)#29.04%
```

```python
#2. Eliminar aquellas que tienen m√°s de un dato faltante:
clean1.dropna(thresh=2, inplace=True) #deja las filas que tienen por lo menos 
#2 valores no nulos (todas tienen por lo menos 2 valores no nulos, entonces queda igual)
clean1.info() #en este caso vemos que todas las que tenias datos 
#faltantes tenian mas de 2 
#Para pensar ü§î: ¬øQu√© desventajas crees que tiene esta forma de lidiar 
#con los datos faltantes?¬øCu√°ndo lo usar√≠as?
#Esta forma de tratar los faltantes introduce sesgo y reduce el tama√±o muestral.
```

```python
#Perdida de datos
#no perdi nada
100 - (len(clean1) * 100) / len(personas) #0.0
100 - (len(clean1.columns)*100) / len(personas.columns) #0.0
```
